## ¿Qué es Python?

## 3 Ejemplos Sencillos de Aplicación

Aquí tienes tres scripts básicos que ilustran cómo Python facilita la vida en la planta o en el campo.

### 1. Monitoreo de Temperatura (Control de Calidad)

Imagina que recibes una señal de un sensor. Este código evalúa si la temperatura de un silo es segura.

```
# Verificación de rango térmico en almacenamiento
temperatura = 28.5  # Dato que vendría del sensor

if temperatura > 25:
    print("ALERTA: Temperatura fuera de rango. Activar refrigeración.")
else:
    print("Estado óptimo: El grano se mantiene estable.")
```

### 2. Cálculo de Rendimiento de Extracción

Este código ayuda a calcular rápidamente el porcentaje de eficiencia de un proceso de extracción de aceite o pulpa.

```
# Cálculo de rendimiento de extracción de aceite
materia_prima_kg = 500
producto_final_kg = 85

rendimiento = (producto_final_kg / materia_prima_kg) * 100

print(f"El rendimiento del proceso es del {rendimiento}%")
```

### 3. Estimación de Riego Basado en Humedad

Un script simple para decidir si se debe activar un sistema de riego por goteo.

```
# Lógica simple para riego automatizado
humedad_suelo = 15  # Porcentaje de humedad

def decidir_riego(humedad):
    if humedad < 20:
        return "ACTIVAR RIEGO: Humedad crítica."
    else:
        return "RIEGO APAGADO: Humedad suficiente."

print(decidir_riego(humedad_suelo))
```
