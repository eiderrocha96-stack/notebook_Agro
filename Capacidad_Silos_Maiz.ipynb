#Capacidad de Silos de Granos

**Contexto:**

Una planta procesadora de maiz requiere calcular cuántas **toneladas de maíz puede almacenar un silo cilíndrico**(tma). El usuario debe ingresa el radio y la altura del silo en metros.

**Requerimiento:**
1. Solicitar el nombre del lote del silo
2. Solicitar el radio(r) del silo en metros (m)
3. Solicitar la altura (h) del silo en metros(m).
4. calcular la capacidad de maiz que se puede almacenar sabiendo que la densidad del maiz es de 0.75 tol/m^3.(tma)

$$V = π(3.1416) * r^2 * h $$  

$$ tma = volumen * 0.75 $$


**Prueba del progrma**

Si se ingresa un silo con  radio de **3 metros** y altura de **10 metros**, el programa debe indicar que la capacidad del silo en toneladas es de **212.06 toneladas** aproximadamente.  




````python
# ENTRADA: solicitar el nombre del lote del silo
# Solicitar el radio(r) del silo en metros (m)
# Solicitar la altura (h) del silo en metros(m).

nombre = input("Ingrese el nombre del lote del silo: ")
r = float(input("Ingrese el radio del silo en metros: "))
h = float(input("Ingrese la altura del silo en metros: "))

# PROCESO: Calcular el volumen(V) y la capacidad (tma) del silo

V = (3.1416) * r**2 * h
tma = round( V * 0.75, 3)

# SALIDA: informar la capacidad del silo

print("La capacidad del silo", nombre, "es de", tma, "toneladas")
````
