# Dosificación de Fertilizante Líquido

**Contexto:**

En una finca, se necesia un **sistema de riego por goteo requiere aplicar una dosis de fertilizante por hectárea**. El usuario ingresa las hectáreas totales y la dosis requerida (litros/hectárea).

**Requerimiento**

1. Solicitar el nombre de la finca
2. Solicitar el numero de hectareas totales(H)
3. Solicitar las dosis requeridas (litros/hectarea)(DR)
4. calcular la disis de fertilizante por hectarea sabiendo que el fertilizante se diluye en agua en una relación 1:500 (1 litro de producto por 500 de agua)(FH)

$$ PT = H * DR $$
$$ DFH = HT * Relacion de Agua (500) $$


**Prueba del programa**

Si se ingresan 5 hectáreas con una Dosis de  2 L/ha, el programa debe indicar que  10L de producto y 5000L de agua total.



````python
#ENTRADA:Solicitar el nombre de la finca
#Solicitar el numero de hectareas totales(H)
#Solicitar las dosis requeridas (litros/hectarea)(DR)

nombre = input("Ingrese el nombre de la finca: ")
hectareas = float(input("Ingrese el numero de hectareas totales: "))
dosis = float(input("Ingrese las dosis requeridas (litros/hectarea): "))

#PROCESO:calcular el agua total (AT) y la disis de fertilizante por hectarea (DFH)
AT = hectareas * dosis
DFT = hectareas * 500

#SALIDA: informar la dosis requerida
print("La dosis requerida para la finca", nombre, "es de", AT, "litros de producto y", DFT, "litros de agua total.")
````
